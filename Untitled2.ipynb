```python
import os
from urllib.request import urlretrieve
import pandas as pd

URL='https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'

def get_fremont_data(filename='Fremont.csv', url=URL, force_download=False ):
    if force_download or not os.path.exists(filename):
        urlretrieve(url, filename)
    data = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
    cols=list(data)
    cols.insert(0,cols.pop(cols.index('Fremont Bridge West Sidewalk')))
    data=data.loc[:,cols]
    data.columns=['West', 'East']
    data['Total']=data['West'] +data['East']
    return data 
```

```python
data = get_fremont_data()
data.head() 
```

```python
data.resample('W').sum().plot();
```

```python
ax = data.resample('D').sum().rolling(365).sum(). plot();
ax.set_ylim(0, None);
```

Each of this points is the sum of the previous 365 days


```python
data.groupby(data.index.time).mean().plot();
```

```python
pivoted=data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.iloc[:5, :5]
```

"legend=False" to have no legend in the graph

```python
pivoted.plot(legend=False, alpha=0.01);
```

```python
X=pivoted.fillna(0).T.values
X.shape
```

```python
from sklearn.decomposition import PCA
X2=PCA(2,svd_solver='full').fit_transform(X)
```

```python
X2.shape
```

```python
plt.scatter(X2[:,0],X2[:,1])
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(2)
gmm.fit(X)
labels=gmm.predict(X)
labels
```

```python
plt.scatter(X2[:,0],X2[:,1], c=labels, cmap='rainbow');
plt.colorbar();
```

```python
pivoted.T[labels==0].T.plot(legend=False, alpha=0.1);
```

```python
pivoted.T[labels==1].T.plot(legend=False, alpha=0.1);
```

```python
dayofweek =pd.DatetimeIndex(pivoted.columns).dayofweek
```

```python
plt.scatter(X2[:,0],X2[:,1], c=dayofweek, cmap='rainbow');
plt.colorbar();
```
